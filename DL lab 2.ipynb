```python
import time
start_time=time.time()
for i in range(0,10000):
    
    print("cube of ",i+1,"is",(i+1)*(i+1)*(i+1))
end_time=time.time()


start_time1=time.time()
i=1
while(i<10000):
    (i)**3
    print("cube of ",i,"is",(i)**3)
    i=i+1
end_time1=time.time()

print("Time for for loop is",end_time-start_time)
print("Time for while loop is",end_time1-start_time1)
```
